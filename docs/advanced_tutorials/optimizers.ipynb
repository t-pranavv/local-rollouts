# Optimizers

This notebook offers a concise, hands‑on introduction to two state‑of‑the‑art optimization algorithms currently available in **PhyAGI**:

- `DAdaptAdam`: a learning‑rate‑free variant of Adam that automatically scales its step size via *D‑Adaptation*.
- `Lion`: a minimalist optimizer that combines sign‑based updates with momentum, discovered through symbolic regression.

For each optimizer you will:

1. Instantiate it with (mostly) default hyper‑parameters.
2. Minimize a 1‑D quadratic objective  

   $$
   f(w)=\frac12\,(w-3)^2
   $$

   for 10 000 steps.
3. Visualize both the loss curve and the parameter trajectory.

**Tip:** Feel free to experiment with different learning rates, objectives, or dimensionalities to see how each optimizer behaves in other scenarios.


````python
import torch
import matplotlib.pyplot as plt

def objective(w):
    return 0.5 * (w - 3) ** 2

def run_optimizer(opt_cls, lr=None, steps=500, **kwargs):
    w = torch.tensor([10.0], dtype=torch.float32, requires_grad=True)
    history = {"loss": [], "w": []}

    opt_kwargs = kwargs.copy()
    if lr is not None:
        opt_kwargs["lr"] = lr

    optimizer = opt_cls([w], **opt_kwargs)
    for _ in range(steps):
        optimizer.zero_grad()
        loss = objective(w)
        loss.backward()
        optimizer.step()
        history["loss"].append(loss.item())
        history["w"].append(w.item())
        
    return history
````

## DAdaptAdam

`DAdaptAdam` (a.k.a. *Adam with D‑Adaptation*) removes the need to tune the global learning rate by maintaining an adaptive scaling factor \(d_t\) that matches the curvature of the loss. In practice it behaves like Adam with a per‑step, automatically chosen learning rate.

- Drop‑in replacement for Adam (same beta and epsilon parameters).  
- No learning rate warmup required.  
- Particularly helpful when training schedules or budgets are unknown.

```python
from phyagi.optimizers.dadaptation import DAdaptAdam
history = run_optimizer(DAdaptAdam, lr=1e-2, steps=10000)
```

The next cell runs the optimizer and plots the training curve.

````python
from phyagi.optimizers.dadaptation import DAdaptAdam

history = run_optimizer(DAdaptAdam, lr=1e-2, steps=10000)

plt.figure()
plt.plot(history["loss"])
plt.yscale("log")
plt.title("DAdaptAdam - Loss over steps")
plt.xlabel("Step")
plt.ylabel("Loss (log scale)")
plt.show()

plt.figure()
plt.plot(history["w"])
plt.title("DAdaptAdam - Parameter trajectory")
plt.xlabel("Step")
plt.ylabel("w")
plt.show()
````

## Lion

`Lion` (“*EvoLved sIgn mOmeNtum*”) is a simple yet surprisingly effective optimizer introduced by Chen *et al.* (2023). Its update rule keeps only the sign of the momentum, making each step computationally cheap and memory light while retaining strong convergence properties.

- **Simplicity:** two lines of PyTorch code.  
- **Speed:** roughly half the memory footprint of Adam/AdamW.  
- **Performance:** matches or exceeds AdamW on many vision and language benchmarks.

```python
from phyagi.optimizers.lion import Lion
history = run_optimizer(Lion, lr=1e-3, steps=10000)
```

Run the cell below to see Lion in action and compare its curve against DAdaptAdam.

````python
from phyagi.optimizers.lion import Lion

history = run_optimizer(Lion, lr=1e-3, steps=10000)

plt.figure()
plt.plot(history["loss"])
plt.yscale("log")
plt.title("Lion - Loss over steps")
plt.xlabel("Step")
plt.ylabel("Loss (log scale)")
plt.show()

plt.figure()
plt.plot(history["w"])
plt.title("Lion - Parameter trajectory")
plt.xlabel("Step")
plt.ylabel("w value")
plt.show()
````
